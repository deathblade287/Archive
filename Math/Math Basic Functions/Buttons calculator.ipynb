```python
# Assignment for Sep 26th 2020 is to make buttons for addition , subtraction , multiplication , division through tkinter

# Steps : 
# Create basic 4 functions
# Take care of previous value - one operation (add , sub , mult , etc) on multiple numbers (more than 2)
# Take care of BODMAS - (Multiple operations (add and subtract , add and mult and div)) on multiple numbers (more than 2)
```

```python
# Importing library(ies)
from tkinter import *
```

```python
ans = 0
root = Tk()
s = 0

root.title("Aviral's Calculator")

def cc(num):
    z = e.get()
    e.delete(0,END)
    e.insert(0,str(z)+str(num))

# Addition
def add():
    global st
    global ans
    n = int(e.get())
    e.delete(0,END)
    st = "a"
    ans = ans+n
# Subtraction
def sub():
    global st
    global ans
    n = int(e.get())
    e.delete(0,END)
    st = "s"
    ans = ans-n
# Multiplication
def mult():
    global st
    global ans
    n = int(e.get())
    e.delete(0,END)
    st = "m"
    ans = ans*n
# Division
def div():
    global st
    global ans
    n = int(e.get())
    e.delete(0,END)
    st = "d"
    ans = ans/n
# Equal to
def equal():
    global ans
    n2=int(e.get())
    e.delete(0,END)
    if (st == "a"):
        s = ans+n2
    e.insert(0,s)
    ans = 0
#   if (+ presses again):
#        n = s

# Vars
e = Entry (root, width = 50,borderwidth = 10)
e.grid(row=0,column=0,columnspan=3,padx=20,pady=20)

# Buttons for no.
b0 = Button(root, text = "0",padx=30,pady=20,command = lambda:cc(0))
b1 = Button(root, text = "1",padx=30,pady=20,command = lambda:cc(1))
b2 = Button(root, text = "2",padx=30,pady=20,command = lambda:cc(2))
b3 = Button(root, text = "3",padx=30,pady=20,command = lambda:cc(3))
b4 = Button(root, text = "4",padx=30,pady=20,command = lambda:cc(4))
b5 = Button(root, text = "5",padx=30,pady=20,command = lambda:cc(5))
b6 = Button(root, text = "6",padx=30,pady=20,command = lambda:cc(6))
b7 = Button(root, text = "7",padx=30,pady=20,command = lambda:cc(7))
b8 = Button(root, text = "8",padx=30,pady=20,command = lambda:cc(8))
b9 = Button(root, text = "9",padx=30,pady=20,command = lambda:cc(9))
# Buttons for operations (sub , add , etc)
button_add = Button(root,text = "+", padx = 30 , pady = 20, command = add)
button_sub = Button(root,text = "-", padx = 30 , pady = 20, command = sub)
button_mult = Button(root,text = "mult", padx = 30 , pady = 20, command = mult)
button_div = Button(root,text = "div", padx = 30 , pady = 20, command = div)
button_equal = Button(root,text = "=", padx = 30 , pady = 20, command = equal)

# All button locations (row/column)
b0.grid(row = 4,column=0)
b1.grid(row = 1,column=0)
b2.grid(row = 1,column=1)
b3.grid(row = 1,column=2)
b4.grid(row = 2,column=0)
b5.grid(row = 2,column=1)
b6.grid(row = 2,column=2)
b7.grid(row = 3,column=0)
b8.grid(row = 3,column=1)
b9.grid(row = 3,column=2)

button_add.grid(row=4,column=1)
button_sub.grid(row=4,column=2)
button_mult.grid(row=5,column=0)
button_div.grid(row=5,column=1)
button_equal.grid(row=5,column=2)

root.mainloop()
```
